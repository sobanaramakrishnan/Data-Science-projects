Explore the distribution of categoricalnvariables like "Country Code","City" and"Cuisines".

```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_columns = ['Country Code', 'City', 'Cuisines']

# Function to plot the distribution of categorical variables
def plot_categorical_distribution(df, column_name):
    plt.figure(figsize=(10, 6))
    value_counts = df[column_name].value_counts()
    sns.barplot(x=value_counts.index, y=value_counts.values, palette="viridis")
    plt.title(f'Distribution of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.xticks(rotation=90)
    plt.show()

# Plot distributions for each categorical variable
for column in categorical_columns:
    plot_categorical_distribution(df, column)
```

Identify the top cuisines and cities with the highest number of restaurants.

```python
def plot_top_categories(df, column_name, top_n=10):
    plt.figure(figsize=(10, 6))
    value_counts = df[column_name].value_counts().head(top_n)
    sns.barplot(x=value_counts.values, y=value_counts.index, palette="viridis")
    plt.title(f'Top {top_n} {column_name} with Highest Number of Restaurants')
    plt.xlabel('Number of Restaurants')
    plt.ylabel(column_name)
    plt.show()

# Identify and plot the top cities with the highest number of restaurants
plot_top_categories(df, 'City')

# Identify and plot the top cuisines with the highest number of restaurants
# Since 'Cuisines' might contain multiple entries separated by commas, we need to split them
df['Cuisines'] = df['Cuisines'].str.split(',')
all_cuisines = df.explode('Cuisines')
all_cuisines['Cuisines'] = all_cuisines['Cuisines'].str.strip()
plot_top_categories(all_cuisines, 'Cuisines')
```
